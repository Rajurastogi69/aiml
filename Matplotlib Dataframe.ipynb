# Bar Graph


```python
x=[2,8,10]
y=[6,15,20]
plt.bar(x,y,width=0.8)
```

```python
# analyse weekly temp data
day=[1,2,3,4,5,6,7]
max_t=[50,45,35,32,47,27,36]
min_t=[45,40,33,30,44,25,34]
avg_t=[45,43,34,31,45,26,35]
```

```python
plt.plot(day, max_t,label="max")
plt.plot(day,min_t,label="min")
plt.plot(day,avg_t,label="avg")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.legend()
plt.grid()
```

```python
# histogram
a=np.array([22,25,23,22,5,6,7,31,32,35,40,45,50,])
bins=[0,10,20,30,40,50]
plt.hist(a,bins)
```

```python
# pie chart 
slices=[8,3,3,12]
subjects=['AIML','RM','IS','AWT',]
cols=['g','b','r','y',]

explodes=(1,0,2,0)
plt.pie(slices,labels=subjects,colors=cols,startangle=30,explode=explodes)
```

# scipy is a python library used to solve scientific and mathematical problems 

```python
import numpy as np
from scipy.integrate import quad
```

```python
def f(x):
    return 2*x+1
```

```python
I,err = quad(f,0,1)
```

```python
print(I)
```

```python
print(err)
```

```python
def f(x):
    return np.cos(x)
```

```python
xlo=0
xhi=np.pi/2
```

```python
I,err=quad(f,xlo,xhi)
```

```python
print(I)
```
