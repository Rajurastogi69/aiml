```python
# Common imports
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Set style for seaborn
sns.set(style="whitegrid")
```

```python
# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['skyblue', 'lightgreen', 'lightcoral', 'gold']
explode = (0.1, 0, 0, 0)  # Emphasize the 1st slice

# Plot
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=140)

# Features
plt.title('Programming Language Popularity')
plt.legend(labels, title="Languages", loc="upper right")
plt.axis('equal')  # Equal aspect ratio ensures that pie is a circle
plt.tight_layout()
plt.show()
```

```python
# Data
data = np.random.randn(1000)

# Plot
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='steelblue', edgecolor='black', alpha=0.7)

# Features
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.axvline(data.mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
 # Data
categories = ['Math', 'Science', 'History', 'English']
scores = [88, 92, 80, 85]

# Plot
plt.figure(figsize=(8, 5))
bars = plt.bar(categories, scores, color='mediumseagreen', edgecolor='black')

# Add value labels on top
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 1, f'{yval}', ha='center')

# Features
plt.title('Student Scores by Subject')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Data
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(20, 200, size=50)
colors = np.random.rand(50)

# Plot
plt.figure(figsize=(8, 5))
scatter = plt.scatter(x, y, s=sizes, c=colors, cmap='viridis', alpha=0.7, edgecolor='k')

# Features
plt.title('Random Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.colorbar(scatter, label='Color Intensity')
plt.tight_layout()
plt.show()
```

```python
# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(x, y1, label='sin(x)', color='blue', linewidth=2, linestyle='-')
plt.plot(x, y2, label='cos(x)', color='red', linewidth=2, linestyle='--')

# Features
plt.title('Sine and Cosine Waves')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.axhline(0, color='black', linewidth=1, linestyle=':')
plt.grid(True)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
