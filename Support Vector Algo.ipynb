```python
# We can also download data from sklearn library
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC #SVC is support vector classifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
irisData=load_iris()
```

```python
irisData.feature_names
```

```python
irisData.target_names
```

```python
df=pd.DataFrame(irisData.data,columns=irisData.feature_names)
```

```python
df.head()
```

```python
#Appending target column
```

```python
df['target']=irisData.target
```

```python
df
```

```python
# 0-> Setosa 1-> Versicolor 2-> Virginica
# We can visualize each class of target (0,1,2)
```

```python
df[df.target==1]
```

```python
# We can add one more column into df as "fname"
# By using Lambda method
```

```python
df['fname']=df.target.apply(lambda x : irisData.target_names[x])
```

```python
df.head()
```

```python
# We will try to plot 3 types of target using matplotlib
# Create 3 diff dataframes for 3 diff types of targets
# ds1=Setosa ds2=Versicolor ds3=Virginica
```

```python
ds1=df[df.target==0]
ds2=df[df.target==1]
ds3=df[df.target==2]
```

```python
ds1.head()
```

```python
ds2.head()
```

```python
ds3.head()
```

```python
pl.scatter(ds1['sepal length (cm)'],ds1['sepal width (cm)'],color='b',marker='d')
pl.scatter(ds2['sepal length (cm)'],ds2['sepal width (cm)'],color='r',marker='o')
pl.scatter(ds3['sepal length (cm)'],ds3['sepal width (cm)'],color='orange',marker='s')
pl.xlabel('sepal length')
pl.ylabel('sepal width')
```

```python
pl.scatter(ds1['petal length (cm)'],ds1['petal width (cm)'],color='b',marker='d')
pl.scatter(ds2['petal length (cm)'],ds2['petal width (cm)'],color='r',marker='o')
pl.scatter(ds3['petal length (cm)'],ds3['petal width (cm)'],color='orange',marker='h')
pl.xlabel('petal length')
pl.ylabel('petal width')
```

```python
# Dropping fNAME column from dataframe
# X->input y->output
```

```python
X=df.drop(['target','fname'],axis='columns')
```

```python
X
```

```python
Y=df.target
```

```python
Y
```

```python
# Splitting data into train and test
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train
```

```python
X_test
```

```python
svmodel=SVC(kernel='linear')
```

```python
svmodel.fit(X_train,Y_train)
```

```python
svmodel.predict([[4.8,3.0,1.4,0.1]])
```

```python
svmodel.predict([[7.9,3.8,6.4,2.0]])
```

```python
Y_predict=svmodel.predict(X_test)
```

```python
accuracy_score(Y_predict,Y_test)
```

```python
confusion_matrix(Y_predict,Y_test)
```
